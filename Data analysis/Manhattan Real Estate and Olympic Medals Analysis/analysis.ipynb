```python
# Imports
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data1 = pd.read_excel('.\Data analysis\Manhattan Real Estate and Olympic Medals Analysis\Datasets\rollingsales_manhattan.xls', header = 4)
data2 = pd.read_csv('.\Data analysis\Manhattan Real Estate and Olympic Medals Analysis\Datasets\olympics.csv')

#Q1.a OK. In part two, removing outliers and 0s could help better analysis.
#Q1.b OK. Removing outliers such as India and China, could lead to a better analysis.
```

### Analyze the Manhattan rolling sales dataset. Where in Manhattan is the most/least expensive real estate located? What is the relationship between the sales price and gross square feet?

```python
# Remove rows with missing values and duplicates
data1.drop_duplicates(inplace=True)
data1.dropna(inplace=True)

# Identify most expensive and least expensive Sale Price
most_expensive_pric = data1['SALE PRICE'].max()
least_expensive_price = data1['SALE PRICE'].min()

# Identify most expensive and least expensive Address
most_expensive_address = data1['ADDRESS'][data1['SALE PRICE'].idxmax()]
least_expensive_address = data1['ADDRESS'][data1['SALE PRICE'].idxmin()]

print(f"Most Expensive Address: {most_expensive_address} With The Sale Price Of: {most_expensive_pric}")
print(f"Least Expensive Address: {least_expensive_address} With The Sale Price Of: {least_expensive_price}")
```

The most expensive address in Manhattan is 390 GREENWICH STREET with the sale price of 1767913114 and the least expensive address is 234 EAST 4TH STREET witht he sale price of 0. There are multiple Adresses with the sale value of 0 so we just choose one of the as the lowest.

```python
# Calculate correlation coefficient between sale price and gross square feet
correlation_coefficient = data1['SALE PRICE'].corr(data1['GROSS SQUARE FEET'])
print("Correlation Coefficient between Sale Price and Gross Square Feet:", correlation_coefficient)

# PLotting a scatter plot with a linear regression line, along with confidence intervals.
plt.figure(figsize=(10, 6))
sns.regplot(data=data1, x='GROSS SQUARE FEET', y='SALE PRICE', scatter_kws={"alpha":0.5})
plt.title('Relationship between Sale Price and Gross Square Feet')
plt.xlabel('Gross Square Feet')
plt.ylabel('Sale Price')
plt.grid(True)
plt.show()
```

A correlation coefficient of approximately 0.5947 suggests a moderately strong positive correlation between sale price and gross square feet in the Manhattan rolling sales dataset. This with the plot we made indicates that as the gross square feet of a property increases, the sale price tends to increase as well, and vice versa.

### Analyze the 2012 Olympic dataset. What can you say about the relationship between a countrys population and the number of medals it wins? What can you say about the relationship between the ratio of female and male counts and the GDP of that country?

```python
# Combining all medals
data2['Total Medals'] = data2['Gold medals'] + data2['Silver medals'] + data2['Bronze medals']

# Calculate the correlation coefficient between population and total medals
correlation_population_medals = data2['2010 population'].corr(data2['Total Medals'])
print("Correlation between Population and Total Medals:", correlation_population_medals)

# PLotting a scatter plot with a linear regression line, along with confidence intervals.
plt.figure(figsize=(10, 6))
sns.regplot(data=data2, x='2010 population', y='Total Medals', scatter_kws={"alpha": 0.5})
plt.title('Regression Analysis: Population vs. Total Medals')
plt.xlabel('Population (2010)')
plt.ylabel('Total Medals')
plt.grid(True)
plt.show()
```

The correlation coefficient of approximately 0.3711 suggests a moderate positive correlation between a country's population and the total number of medals it wins in the Olympics. This with the plot we made shows that there's a tendency for countries with larger populations to win more medals.

```python
# Calculate female to male ratio
data2['Female/Male Ratio'] = data2['Female count'] / data2['Male count']

# Calculate the correlation coefficient between female/male ratio and GDP
correlation_female_male_gdp = data2['Female/Male Ratio'].corr(data2['2011 GDP'])
print("Correlation between Female/Male Ratio and GDP:", correlation_female_male_gdp)

# PLotting a scatter plot with a linear regression line, along with confidence intervals.
plt.figure(figsize=(10, 6))
sns.regplot(data=data2, x='Female/Male Ratio', y='2011 GDP', scatter_kws={"alpha": 0.5})
plt.title('Regression Analysis: Female/Male Ratio vs. GDP')
plt.xlabel('Female/Male Ratio')
plt.ylabel('GDP (2011)')
plt.grid(True)
plt.show()
```

The correlation coefficient of approximately 0.1737 suggests a weak positive correlation between the female-to-male ratio and GDP. This with the scatter plot we made 
indicates that there might be a slight tendency for countries with higher GDPs to have a slightly higher female-to-male ratio.

**time series** is a sequence of data points collected over time, like daily temperatures or monthly sales figures.

**Trend Effects:** These show the long-term direction the data is moving. For example, increasing sales over several years.

**Seasonal Variations:** These are regular, repeating patterns, like higher sales during holidays or warmer temperatures in summer.

**Random Error:** This is the unpredictable "noise" in the data that doesn't follow any pattern, caused by things like measurement mistakes or chance fluctuations.
